```python
df_cleaned = movies.dropna()
df_cleaned.isna().sum()
```

```python
movies.dropna(subset=['star_rating']).isna().sum()
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
        فیلم ها را براساس مدت زمان فیلم به صورت نزولی مرتب کنید و پنجاه تای اول را در فایل csv با نام long_time ذخیره کنید 
        <br/>
    </font>
</div>

```python
df_sorted = df_cleaned.sort_values(by='duration', ascending=False).head(50)
df_sorted.to_csv('data/long_time.csv', index=False)
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
         تابعی بنویسید که نام بازیگر را به عنوان آرگومان بگیرد و تعداد فیلم های برتر از این لیست را که بازی کرده محاسبه کند.
        <br/>
    </font>
</div>

```python
def count_movie_by_actor(df, actor_name):
    return df['actors_list'].apply(lambda x: actor_name in x).sum()

actor_name = 'Al Pacino'
count = count_movie_by_actor(df_cleaned, actor_name)
print(f'{actor_name} has acted in {count} top movies')
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">

        <br/>
    </font>

```python
df_cleaned.groupby('genre')['star_rating'].agg(['mean', 'count'])
```

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=4>
       ستونی به این داده ها اضافه کنید که در صورت اینکه یک فیلم مدت نمایش آن 150 دقیقه و یا بیشتر بود مقدار صفر و در غیر اینصورت 1 در هر سطر این ستون قرار گیرد. دیتافریم حاصل را در یک فایل csv ذخیره کنید.
        <br/>
    </font>
</div>

```python
df_cleaned['short_movie'] = df_cleaned['duration'].apply(lambda x: 0 if x >=150 else 1)
df_cleaned.to_csv('data/imdb_movie_with_short_flag.csv')
```
