```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = list(range(len(variance)))
```

```python
xs
```

```python
plt.figure(figsize=(6, 2))

plt.plot(xs, variance, 'g-', label='variance') # green solid line
plt.plot(xs, bias_squared, 'r-', label='bias')
plt.plot(xs, total_error, 'b:', label='total_error')

plt.legend(loc=0)
plt.title("The bias-variance Tradeoff")
plt.xlabel("model complexity")
plt.ylabel("error")
plt.xticks(range(0,8, 2))
plt.grid()
plt.show()
```

## Subplots

### 1. Creating Subplots using plt.subplot()

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.subplot(1, 2, 1)
plt.plot(x, y1, color='blue')
plt.title('Sine Wave')


plt.subplot(1, 2, 2)
plt.plot(x, y1, color='red')
plt.title('Cosine Wave')

plt.tight_layout()
plt.show()
```

### 2. Creating Subplots using plt.subplots() (Preferred)

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

axes[0, 0].plot(x, y1)
axes[0, 0].set_title("Sine Wave")


axes[0, 1].plot(x, y2, color='red')
axes[0, 1].set_title("Cosine Wave")


axes[1, 0].plot(x, y3)
axes[1, 0].set_title("Tangent Wave")

axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)

for ax in axes.flatten():
    ax.hist(np.random.randn(100), bins=20, edgecolor='black')

plt.show()
```
