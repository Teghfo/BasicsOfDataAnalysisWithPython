```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
iris_df = sns.load_dataset('iris')
iris_df.head()
```

## Categorical Plots

### 1. barplot === args (x, y, data ? df, estimator)

```python
# plt.bar(tips_df.groupby('smoker')['total_bill'].mean().index, tips_df.groupby('smoker')['total_bill'].mean().values)

sns.barplot(x = 'smoker', y='total_bill', hue='sex', data=tips_df, estimator=np.mean, palette='coolwarm')
```

### 2. countplot

```python
sns.countplot(y='sex', data=tips_df)
```

### 3. boxplot

```python
sns.boxplot(x='total_bill', y='smoker', hue='sex', orient='h',data=tips_df, palette='viridis')
```

### 4. Violinplot

```python
sns.violinplot(x='day', y='tip', hue='sex', data=tips_df)
```

## Distributaion Plots

### 1. distplot

```python
sns.displot(tips_df['total_bill'], kde=True, bins=20)
```

```python
sns.histplot(tips_df['total_bill'], bins=20)
```

```python
sns.kdeplot(tips_df['total_bill'], fill=True)
```

### 2. jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips_df, kind='kde')
```

### 3. pairplot

```python
sns.pairplot(tips_df, hue='sex')
```

```python
sns.pairplot(iris_df, hue='species')
```

### 4. rugplot

```python
sns.rugplot(tips_df['total_bill'])
```

## Correlation

```python
tips_df[['total_bill', 'tip']].corr()
```

```python
sns.heatmap(tips_df[['total_bill', 'tip']].corr(), annot=True, cmap='coolwarm')
```
